```python
hero_info = pd.read_csv('王者荣耀.csv')
hero_info
```

```python
zhanshi_data =hero_info[hero_info['英雄定位']==1]
zhanshi_live = np.array(zhanshi_data['生存能力'])
zhanshi_damage = np.array(zhanshi_data['攻击伤害'])
zhanshi_skill = np.array(zhanshi_data['技能效果'])
```

```python
def plot3D(x,y,z):
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.scatter(x, y, z, c=z, linewidth=0.5)
    plt.show()
    pass

def plot2D(x,y,color):
    plt.scatter(x,y,color=color)
    plt.show()
    pass


```

```python
plot3D(zhanshi_damage,zhanshi_live,zhanshi_skill)
plot2D(zhanshi_live,zhanshi_damage,'r')
```
